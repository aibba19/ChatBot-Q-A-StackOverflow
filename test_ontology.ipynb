```python
def calculate_median(queries,tech):
    
    similarity_scores = []
    
    for q in queries:
        
        if tech == 'w2v':
            results = w2v_ontology_result(q)
        else:
            results =  tfidf_onto_result(q)
        #creiamo una lista con solo i valori di similarit√† di ogni primo risultato per query
        sim_value = results[0][2]
        
        similarity_scores.append(sim_value)
    
    median = statistics.median(similarity_scores)
    
    return median 
```

```python
def create_test_table(queries, tech, median):
    
    res_ground_truth = []
    
    for q in queries:
        if tech == 'w2v':
            results = w2v_ontology_result(q)
        else:
            results = tfidf_onto_result(q)
        i = 0 
        
        res_q = [0,0,0]
        
        for r in results:
            sim_value = r[2]
            
            if sim_value >= median:
                res_q[i] = 1
            else:
                res_q[i] = 0
                    
            i = i+1
        res_ground_truth.append(res_q)
        
    data = res_ground_truth 
    
    #print(data)
    test_df = pd.DataFrame.from_records(data)
 
    # specifying column names
    test_df.columns = ['Res1', 'Res2', 'Res3']
    
    test_df = test_df.assign(queries=pd.Series(queries).values)
    
    return test_df
```

```python
def create_evaluation_table(test_results, test_evaluated):
    data = []
    #Ciclo sul df che contiene i risultati del modello 
    for index,row in test_results.iterrows():
        #Lista di risultati valutati personalmente
        r_evaluated = test_evaluated.iloc[index]
        #print('r_evaluated:' , r_evaluated)
        
        final_res = ['','','']
        i = 1
        for r in row:
            if (isinstance(r, int)):
                #print('r_evaluated[i]: ', r_evaluated[i] )
                #Se r √® 1 vuol dire che per il modello il record √® stato registrato come rilevante
                if r == 1:
                    if r == r_evaluated[i]:           
                        #True positive
                        final_res[i-1] = 'tp'
                    else:
                        #False positive
                        final_res[i-1] = 'fp'
                else:
                    if r == r_evaluated[i]:
                        #True negative
                        final_res[i-1] = 'tn'
                    else:
                        #False negative
                        final_res[i-1] = 'fn'

                i = i+1
        data.append(final_res)
        
    #print(data)
    final_df = pd.DataFrame.from_records(data)
 
    # specifying column names
    final_df.columns = ['Res1', 'Res2', 'Res3']
    
    final_df = final_df.assign(queries=pd.Series(queries).values)
    
    return final_df
```

```python
def create_confusion_matrix(test_evaluation):
    
    tp = 0 
    fp = 0
    tn = 0
    fn = 0 
    
    for rowIndex, row in test_evaluation.iterrows(): #iterate over rows
        for columnIndex, value in row.items():
            if value == 'tp':
                tp = tp + 1
            if value == 'fp':
                fp = fp + 1
            if value == 'tn':
                tn = tn + 1
            if value == 'fn':
                fn = fn  + 1
                
    d = {'Relevant': [tp, fn], 'Non_relevant': [fp, tn]}
    
    confusion_matrix = pd.DataFrame(data = d, index=['Retrieved','Not_retrieved'])
    
    precision = tp/ (tp+fp)
    #print('prec: ',precision)
    recall = tp / (tp+fn)
    #print('recall: ', recall)
    #f1 calcolato con valore parametri beta = 1 e alpha = 1/2
    f1score = (2*precision*recall) / (precision+recall)
    
    return (confusion_matrix,precision,recall,f1score)
    
```

```python
#funzione che prende in ingresso una riga con valori ad 1 dove √® stato trovato un match
#e NaN dove no
#Calcolo l'average precision per ciascuna riga
def average_precision(row):
    
    #res1 = row['tfidf_res1']
    #res2 = row['tfidf_res2']
    #res3 = row['tfidf_res3']
    
    res1 = row.iloc[0]
    res2 = row.iloc[1]
    res3 = row.iloc[2]
        
    if math.isnan(res1) and math.isnan(res2) and math.isnan(res3):
        
        
        return 0
    
    else:
        #La i tiene conto dei risultati rilevanti
        i = 1
        #la j tiene conto del totale dei risultati esaminati 
        j = 1
        numerator = 0 
        
        for el in row:
            if el == 1:
                numerator = numerator + (i/j)
                i = i+1
            j = j+1
            
        avg_precision = numerator/i
        
        return avg_precision
                
                
        
```

```python
#Funzione che calcola il valore della metrica map al quale passare le colonne relative ai test in ordine di indice
def calculate_map(test_results):
 
    #test_results = test_results[['tfidf_res1', 'tfidf_res2','tfidf_res3']]
    avg_prec = 0 
    n_test = len(test_results)
    
    print('Number of test record:' + str(n_test) + '\n')
    
    for index, row in test_results.iterrows():
        avg_prec = avg_prec + average_precision(row)
        
        
    map_value = avg_prec/n_test
    
    return map_value
    
```

```python
def calculate_precision(test_results):
    
    prec_value = 0 
    res_matched = 0
    n_returned = 0
    
    n_test = len(test_results)
    
    print('Number of test record:' + str(n_test) + '\n')
    
    for index,row in test_results.iterrows():
        
        
        #print('Number of results for each row:' + str(n_returned))
        
        
        
        #print(row)
        
        for el in row:
            
            n_returned = n_returned + 1
            
            if el == 1:
                res_matched = res_matched + 1
                #print('Number of res matched: ' + str(res_matched) + '\n')
                
                
        #prec_value = prec_value + (res_matched/n_returned)
    
    prec_value = res_matched/n_returned
    
    return prec_value
```

```python
def calculate_recall(test_results):
    
    recall_value = 0
    res_matched = 0 
    #Questo √® uguale ai risultati relativi per ogni query visto che ogni query ha una funzione da rendere 
    #Perch√® per ogni query vogliamo una ed una sola funzione 
    n_test = len(test_results)
    
    for index,row in test_results.iterrows():
        
        
        #print('Number of results for each row:' + str(n_returned))
        
        
        
        #print(row)
        
        for el in row:
            
            if el == 1:
                res_matched = res_matched + 1
        
    
    recall_value = res_matched/n_test
    
    return recall_value
    
    
    
```
